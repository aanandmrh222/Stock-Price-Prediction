```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Assuming you have already split your data into X_train and y_train

# If X_train has more than 2 dimensions, reshape it
# For example, if X_train is of shape (n_samples, n_timesteps, n_features)
# You can reshape it to (n_samples, n_timesteps * n_features)
X_train_2d = X_train.reshape(X_train.shape[0], -1)

# Train a linear regression model
model = LinearRegression()
model.fit(X_train_2d, y_train)

# Make predictions on the training data
y_pred = model.predict(X_train_2d)

# Calculate mean squared error (optional)
mse = mean_squared_error(y_train, y_pred)
print("Mean Squared Error:", mse)

# Plot actual vs predicted
plt.figure(figsize=(10, 6))
plt.scatter(y_train, y_pred, color='blue', label='Actual vs Predicted')
plt.plot([y_train.min(), y_train.max()], [y_train.min(), y_train.max()], 'k--', lw=3, label='Perfect Prediction')
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.title('Actual vs Predicted Stock Prices')
plt.legend()
# plt.grid(True)
plt.show()
```
